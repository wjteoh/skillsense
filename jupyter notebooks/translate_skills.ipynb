```python
non_EN_skills = DATA_DIR + 'more_NON-EN.txt'
# with open(non_EN_skills, 'r') as f:
#     non_EN_skills_list = f.read()
    
with open(non_EN_skills, 'r') as f:
    content = f.readlines()
# you may also want to remove whitespace characters like `\n` at the end of each line
non_EN_skills_list = [x.strip() for x in content] 
# non_EN_skills_list
non_EN_skills_list
```

```python
result = translate_client.translate(non_EN_skills_list[:50])
converted = []
for i, res_i in enumerate(result):
#     print i
#     print u'Text:', res_i['input']
    print u'Translation:', res_i['translatedText']
    converted.append([unicode(res_i['translatedText']).encode('utf-8')])
    
translated_df = pd.DataFrame(converted, columns=['translated'])
translated_df.to_csv(DATA_DIR + 'dict_skill_translated2.csv', index=False)
```
